1. Dengan menggunakan figsize, perhatikan apakah ukuran image pixelnya juga berubah? Tidak, ukuran pixel dari gambar tidak berubah ketika mengubah ukuran gambar menggunakan parameter figsize pada library matplotlib di Python. Parameter figsize digunakan untuk mengatur ukuran dimensi gambar dalam satuan inch (inci) pada saat dicetak atau ditampilkan, sementara ukuran pixel dari gambar ditentukan oleh resolusi gambar yang telah ditentukan sebelumnya. Oleh karena itu, meskipun mengubah ukuran gambar menggunakan figsize, ukuran pixel dari gambar tetap sama dan hanya akan mempengaruhi ukuran tampilan gambar dalam satuan inch. Untuk mengubah ukuran pixel dari gambar, kita dapat menggunakan parameter dpi (dots per inch) atau resolution pada saat menyimpan atau mengekspor gambar. Atau, kita dapat menggunakan library lain seperti Pillow atau OpenCV untuk memanipulasi ukuran pixel dari gambar secara langsung.


```python
#2

imgt2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Select Red-Blue channel
img_rb = np.zeros_like(imgt2)
img_rb[:,:,0] = img[:,:,0] # Red channel
img_rb[:,:,2] = img[:,:,2] # Blue channel

# Select Green-Blue channel
img_gb = np.zeros_like(imgt2)
img_gb[:,:,1] = img[:,:,1] # Green channel
img_gb[:,:,2] = img[:,:,2] # Blue channel

# Display images
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img_rb)
plt.title("Red-Blue channel")
plt.subplot(1,2,2)
plt.imshow(img_gb)
plt.title("Green-Blue channel")
plt.show()
```
